```python
# Import libraries
print "Importing libraries"

import numpy as np 
import matplotlib.pyplot as plt
import tensorflow as tf 
import keras
from keras.models import Sequential
from keras.layers import Input,Dense, Dropout, Flatten
import pydot
import graphviz
import time

print "Libraries imported correctly."
print "Loading data..."

# GET DATA
from reader import get_images
(x_train, y_train), (x_test, y_test) = get_images()
num_classes = 10
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)

print "Data loaded successfully."
print "==================================="
print ""
print ""

# DEFINE PARAMETERS
nInput = [125]
lossFunct = ["categorical_crossentropy"]
opt = ["sgd"]
metric = ["accuracy"]
batchSize = [275, 300, 325]
epoch = [100]

# DEFINE DATA STRUCTURE TO HOLD OPTIMAL COMBINATION
bestParams = [None, None, None, None, None, None]
bestScenario = (bestParams, 0)

#COUNT TOTAL NUMBER OF ITERATIONS
iterationTotal = len(nInput) * len(lossFunct) * len(opt) * len(metric) * len(batchSize) * len(epoch)

startTime = time.time()

# ITERATE THROUGH ALL PARAMS COMBINATIONS
iterationCount = 0
for nI in nInput:
    for lF in lossFunct:
        for oP in opt:
            for mT in metric:
                for bS in batchSize:
                    for eP in epoch:
                        iterationCount += 1
                        print "Testing combination ", iterationCount, " of ", iterationTotal
                        # BUILD MODEL
                        model = keras.models.Sequential()
                        model.add(Dense(nI, input_shape=(784,), activation='sigmoid'))
                        model.add(Dense(10, activation='softmax'))
                        model.compile(loss=lF, optimizer=oP, metrics=[mT])
                        # TRAIN MODEL
                        batch_size = bS
                        epochs = eP
                        model.fit(x_train, y_train, batch_size, epochs, verbose=0)
                        # EVALUATE ACCURACY
                        score = model.evaluate(x_test, y_test, verbose=0)
                        # COMPARE WITH BEST SOLUTION SO FAR
                        if score[1] > bestScenario[1]:
                            bestParams = [nI, lF, oP, mT, bS, eP]
                            bestScenario = (bestParams, score[1])
                            print "======> Found better combination, accuracy: ", score[1]
                            print "======> Partameters: ", bestParams
                        timeElapsed = time.time() - startTime
                        print "Time elapsed: ", int(timeElapsed/3600),":",int(timeElapsed/60), ":", int(timeElapsed%60)
                            
# TELL USER BEST COMBINATION
print ""
print ""
print "The best performance of ", bestScenario[1], " is acheived with the parameters:"
print bestScenario[0]
```
